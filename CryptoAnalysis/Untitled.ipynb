```python
def get_daily_change(df):
    
    df['daily_percentage'] = df['close']
    for i in range(0,len(df['close']+1)):

        if df['open'][i] > df['close'][i]:
            
            df['daily_percentage'][i] = ((df['close'][i] - df['open'][i])*100)/df['open'][i]
            
        if df['open'][i] < df['close'][i]:
            
            df['daily_percentage'][i] = ((df['close'][i] - df['open'][i])*100)/df['open'][i]
        
```
